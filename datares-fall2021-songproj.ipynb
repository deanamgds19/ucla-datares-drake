```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
drake = pd.read_csv("Drake.csv")
```

```python
drake
```

```python
# songs with most similar lyrics get clustered together
lyric_list = list(drake["Lyric"])
drake_sayings = {"more life": 0, "6 god": 0, "views": 0, "ovo": 0, "catch up": 0, 
                 "on one": 0, "no one else": 0, "award":0, "i know":0, "i need":0}

for song in lyric_list:
    for saying in drake_sayings.keys():
        if saying in str(song):
            drake_sayings[saying] += 1
```

```python
drake_sayings
```

```python
keys = drake_sayings.keys()
values = drake_sayings.values()
plt.bar(keys, values)
plt.xticks(rotation = 20)
plt.xlabel("Drake Saying")
plt.ylabel("Number of Songs")
plt.suptitle("Count of Frequent Drake Sayings")
plt.savefig("drake_sayings.png")
```
